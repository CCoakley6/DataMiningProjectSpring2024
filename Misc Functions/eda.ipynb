```python
import pandas as pd
import glob
```

```python
# Path to the directory where your CSV files are located
directory_path = "../Test CSVs/"

# Use glob to get all the CSV files in the directory
csv_files = glob.glob(directory_path + '*.csv')

# Initialize an empty DataFrame
full_df = pd.DataFrame()

# Loop over the list of csv files
for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Append the contents of the file to the full DataFrame
    full_df = pd.concat([full_df, df], ignore_index=True)

# Reset the index of the final DataFrame (optional)
full_df.reset_index(drop=True, inplace=True)

# Now `full_df` contains all the data from the CSV files in the directory
```
