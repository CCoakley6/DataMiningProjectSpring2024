```python
import pandas as pd
import glob
import numpy as np

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)
```

```python
columns_to_load = [
     'location_key'
    , 'date'
    , 'place_id'
    , 'wikidata_id'
    , 'datacommons_id'
    , 'country_code'
    , 'country_name'
    , 'subregion1_code'
    , 'subregion1_name'
    , 'subregion2_code'
    , 'subregion2_name'
    , 'new_confirmed'
    , 'new_deceased'
    , 'cumulative_confirmed'
    , 'cumulative_deceased'
    , 'new_persons_fully_vaccinated'
    , 'cumulative_persons_fully_vaccinated'
    , 'population'
    , 'population_male'
    , 'population_female'
    , 'population_age_00_09'
    , 'population_age_10_19'
    , 'population_age_20_29'
    , 'population_age_30_39'
    , 'population_age_40_49'
    , 'population_age_50_59'
    , 'population_age_60_69'
    , 'population_age_70_79'
    , 'population_age_80_and_older'
    , 'openstreetmap_id'
    , 'latitude'
    , 'longitude'
    , 'area_sq_km'
    , 'life_expectancy'
    , 'mobility_retail_and_recreation'
    , 'mobility_grocery_and_pharmacy'
    , 'mobility_parks'
    , 'mobility_transit_stations'
    , 'mobility_workplaces'
    , 'mobility_residential'
    , 'average_temperature_celsius'
    , 'minimum_temperature_celsius'
    , 'maximum_temperature_celsius'
    , 'rainfall_mm'
    , 'dew_point'
    , 'relative_humidity'
]
```

```python
# Directory path to your CSV files
directory_path = "../All CSVs/"  # Example directory path

# Initialize an empty DataFrame to store data from all CSV files
full_df = pd.DataFrame(columns=columns_to_load)

# Loop over the list of CSV files
for file in glob.glob(directory_path + '*.csv'):
    # Read the CSV file without specifying columns to ensure we don't miss any data
    df = pd.read_csv(file)
    
    # Ensure all desired columns are present, add them with NaN values if they are missing
    for column in columns_to_load:
        if column not in df.columns:
            df[column] = np.nan

    # Reorder and select only the desired columns to maintain consistency
    df = df[columns_to_load]
    
    # Append the contents of the file to the full DataFrame
    full_df = pd.concat([full_df, df], ignore_index=True)

# Optionally, reset the index of the final DataFrame
full_df.reset_index(drop=True, inplace=True)

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

cluster1_df = full_df.copy()

# Data Preparation
# Using direct column assignment to avoid future compatibility issues
cluster1_df['vaccination_rate'] = cluster1_df['cumulative_persons_fully_vaccinated'] / cluster1_df['population']
cluster1_df['death_rate'] = cluster1_df['new_deceased'] / cluster1_df['new_confirmed']
cluster1_df['confirmed_rate'] = cluster1_df['new_confirmed'] / cluster1_df['population']
cluster1_df['old_pop'] = (cluster1_df['population_age_50_59'] + cluster1_df['population_age_60_69']) / cluster1_df['population']

# Convert columns to numeric, ensuring there are no non-numeric types
numeric_columns = ['vaccination_rate', 'death_rate', 'confirmed_rate', 'old_pop']
for col in numeric_columns:
    cluster1_df[col] = pd.to_numeric(cluster1_df[col], errors='coerce')

# Drop NaN values safely
cluster1_df = cluster1_df.dropna(subset=numeric_columns)

# Ensure all rates are finite
cluster1_df = cluster1_df[np.isfinite(cluster1_df[numeric_columns]).all(axis=1)]

# Filter tiny rates
cluster1_df = cluster1_df[(cluster1_df['death_rate'] > 0.0000001) & (cluster1_df['confirmed_rate'] > 0.000001)]

# Define your threshold for 'new_deceased' to create a binary target
death_rate_threshold = cluster1_df['death_rate'].median()
cluster1_df['high_death_rate'] = (cluster1_df['death_rate'] > death_rate_threshold).astype(int)

# Select features and target variable
features = ['vaccination_rate', 'old_pop']
target = 'high_death_rate'

# Split the dataset
X = cluster1_df[features]
y = cluster1_df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Building the Decision Tree
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)

# Predict and Evaluate the Model
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Plot the Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=features, class_names=['Low', 'High'], filled=True)
plt.savefig('tree_high_res.png')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
cluster2_df = full_df.copy()

# Load your data - assuming it's already loaded as cluster2_df

# Step 1: Data Preparation
# Filter out necessary columns and calculate new_confirmed per capita
cluster2_df['confirmed_per_capita'] = cluster2_df['new_confirmed'] / cluster2_df['population']
data = cluster2_df[['average_temperature_celsius', 'relative_humidity', 'confirmed_per_capita']]
data.dropna(inplace=True)  # Remove missing values

# Filter out small ratios, e.g., where confirmed_per_capita is less than 0.0001
data = data[data['confirmed_per_capita'] >= 0.0001]

# Step 2: Feature Selection is already done in Step 1 by filtering columns

# Step 3: Scaling the Data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 4: Clustering
# Choosing the number of clusters with the elbow method
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_scaled)
    inertias.append(kmeans.inertia_)

# Plotting the elbow plot
plt.figure(figsize=(8, 4))
plt.plot(range(1, 10), inertias, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Let's assume the elbow point is at k=3
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(data_scaled)

# Step 5: Analysis
data['Cluster'] = clusters

# Step 6: Visualization
# 3D Scatter plot of Temperature vs. Humidity vs. Confirmed Cases per Capita colored by cluster
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data['average_temperature_celsius'],
                     data['relative_humidity'],
                     data['confirmed_per_capita'],
                     c=data['Cluster'], cmap='viridis', s=100, alpha=0.6)
ax.set_xlabel('Average Temperature (Celsius)')
ax.set_ylabel('Relative Humidity (%)')
ax.set_zlabel('Confirmed Cases per Capita')
ax.set_title('3D Clusters by Temperature, Humidity, and COVID-19 Cases per Capita')
plt.legend(*scatter.legend_elements(), title='Cluster')
plt.show()

# Plot for Confirmed Cases per Capita by Cluster
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='confirmed_per_capita', data=data)
plt.title('Confirmed Cases per Capita by Cluster')
plt.ylabel('Confirmed Cases per Capita')
plt.xlabel('Cluster')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Load your data - assuming it's already loaded as cluster2_df

# Step 1: Data Preparation
# Filter out necessary columns and calculate new_confirmed per capita
cluster2_df['confirmed_per_capita'] = cluster2_df['new_confirmed'] / cluster2_df['population']
data = cluster2_df[['average_temperature_celsius', 'relative_humidity', 'confirmed_per_capita', 'new_confirmed']]
data.dropna(inplace=True)  # Remove missing values

# Filter out small ratios, e.g., where confirmed_per_capita is less than 0.0001
data = data[data['confirmed_per_capita'] >= 0.0001]

# Step 2: Feature Selection is already done in Step 1 by filtering columns

# Step 3: Scaling the Data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['average_temperature_celsius', 'relative_humidity', 'confirmed_per_capita']])

# Step 4: Clustering
# Choosing the number of clusters with the elbow method
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_scaled)
    inertias.append(kmeans.inertia_)

# Plotting the elbow plot
plt.figure(figsize=(8, 4))
plt.plot(range(1, 10), inertias, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Let's assume the elbow point is at k=3
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(data_scaled)

# Step 5: Analysis
data['Cluster'] = clusters

# Visualization: Total Confirmed Cases by Cluster
cluster_confirmed_totals = data.groupby('Cluster')['new_confirmed'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Cluster', y='new_confirmed', data=cluster_confirmed_totals)
plt.title('Total Confirmed Cases by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Total Confirmed Cases')
plt.show()
```
