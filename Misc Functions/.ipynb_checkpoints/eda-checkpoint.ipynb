```python
import pandas as pd
import glob

# Path to the directory where your CSV files are located
directory_path = "../Single State/"

# List of columns to keep
columns_to_keep = [
    'location_key', 'date', 'place_id', 'subregion1_code', 'subregion1_name',
    'subregion2_code', 'subregion2_name', 'new_confirmed', 'new_deceased',
    'cumulative_confirmed', 'cumulative_deceased', 'new_persons_fully_vaccinated',
    'cumulative_persons_fully_vaccinated', 'population', 'population_male',
    'population_female', 'population_age_00_09', 'population_age_10_19',
    'population_age_20_29', 'population_age_30_39', 'population_age_40_49',
    'population_age_50_59', 'population_age_60_69', 'population_age_70_79',
    'population_age_80_and_older', 'openstreetmap_id', 'latitude', 'longitude',
    'area_sq_km', 'life_expectancy', 'mobility_retail_and_recreation',
    'mobility_grocery_and_pharmacy', 'mobility_parks', 'mobility_transit_stations',
    'mobility_workplaces', 'mobility_residential', 'average_temperature_celsius',
    'minimum_temperature_celsius', 'maximum_temperature_celsius', 'rainfall_mm',
    'dew_point', 'relative_humidity'
]

# Use glob to get all the CSV files in the directory
csv_files = glob.glob(directory_path + '*.csv')

# Initialize an empty DataFrame
full_df = pd.DataFrame()

# Loop over the list of csv files
for file in csv_files:
    # Read the CSV file without specifying usecols
    df = pd.read_csv(file)
    
    # Check for missing columns and add them with default value of 0
    missing_columns = [col for col in columns_to_keep if col not in df.columns]
    for col in missing_columns:
        df[col] = 0
    
    # Reorder and keep only the specified columns
    df = df[columns_to_keep]
    
    # Append the contents of the file to the full DataFrame
    full_df = pd.concat([full_df, df], ignore_index=True)

# Reset the index of the final DataFrame (optional)
full_df.reset_index(drop=True, inplace=True)

# Display the shape of the final DataFrame
print(full_df.shape)
```

```python
column_null_counts = df.isnull().sum()
total_rows = len(df)
percentage_null = (column_null_counts / total_rows) * 100

# Filter columns where the percentage of non-null values is more than 60%
columns_to_keep = percentage_null[percentage_null < 40].index

# Create a new DataFrame with only the selected columns
filtered_df = df[columns_to_keep]

filtered_df
```

```python
import pandas as pd

def get_fill_ratios(dataframe):
    """
    Calculates the fill ratios (percentage of non-null values) for each column in the DataFrame.

    Parameters:
    dataframe (pandas.DataFrame): The DataFrame to calculate fill ratios for.

    Returns:
    pandas.Series: A Series containing fill ratios for each column.
    """
    total_rows = len(dataframe)
    column_null_counts = dataframe.isnull().sum()
    return (1 - column_null_counts / total_rows) * 100

# Example DataFrame



# Calculate fill ratios for each column
fill_ratios = get_fill_ratios(df)

# Write column names and fill ratios to Excel
with pd.ExcelWriter('fill_ratios.xlsx') as writer:
    fill_ratios_df = pd.DataFrame({'Column': fill_ratios.index, 'Fill Ratio (%)': fill_ratios.values})
    fill_ratios_df.to_excel(writer, index=False)
```
